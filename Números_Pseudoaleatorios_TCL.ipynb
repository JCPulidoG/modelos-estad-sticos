## TCL Visualmente


```python
import pandas as pd
import seaborn as sns
```

```python
df = pd.DataFrame()

for i in range (1, 500):
  muestra = np.random.binomial(1000, 0.6, 10) # muestra el # de éxitos con 1000 ensayos Bernoulli
  col = f"muestra {i}"
  df[col] = muestra

df.head()
```

```python
df_muestra_medias = pd.DataFrame(df.mean(), columns = ["Muestra de medias"])
df_muestra_medias
```

```python
sns.distplot(df_muestra_medias)
```
